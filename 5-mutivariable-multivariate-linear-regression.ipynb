```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('/kaggle/input/melbourne/Melbourne_housing_FULL.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
# First remove the rows which have y as nan
df = df.dropna(subset=['Price'])
```

```python
df
```

```python
X = df.drop(columns=['Price', 'Address', 'Date', 'YearBuilt', 'Lattitude', 'Longtitude'])
```

```python
Y = df[['Price']]
```

```python
Y.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3)
```

```python
num_cols = X_train.select_dtypes(include=['int64', 'float64']).columns
```

```python
cat_cols = X_train.select_dtypes(include=['object']).columns
```

```python
num_cols
```

```python
cat_cols
```

```python
# Now apply the simpleImputer to all
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
num_pipe = Pipeline([
    ('simpleImputer', SimpleImputer(strategy='mean')),
    ('scale', StandardScaler())
])

cat_pipe = Pipeline([
    ('simpleImputer', SimpleImputer(strategy='most_frequent')),
    ('oneHot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
transformer = ColumnTransformer(
    transformers = [
    ('NumericalHandling', num_pipe, ['Rooms', 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea', 'Propertycount']),
    ('category', cat_pipe, ['Suburb', 'Type', 'Method', 'SellerG', 'CouncilArea', 'Regionname'])
    ],
    remainder = 'passthrough'
)
```

```python
transformer.fit(X_train)
```

```python
X_train_trans = transformer.transform(X_train)
X_test_trans = transformer.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train_trans, y_train)
```

```python
reg.score(X_train_trans, y_train)
```

```python
reg.score(X_test_trans, y_test)
#this .score automaticaly gives the r2_score
# Proof can be seen below
```

```python
from sklearn.metrics import r2_score

print(r2_score(y_train, reg.predict(X_train_trans)))
print(r2_score(y_test, reg.predict(X_test_trans)))
```

```python
# Seeing there is not underfitting, overfitting
# But lets continue with the Next methods, for understanding purpose,
# Some times your training score will be higher and testing score will be lower its called Overfitting

# One example
# Here training score is 68% but test score is 13.85% which is very low
# Normal Regression is clearly overfitting the data, let's try other models
# Some times your training score itself will be very low, its called underFitting
```

LETS Apply the Regularization Techniques
1. Ridge Regression
2. Laso Regression

```python
from sklearn.linear_model import Ridge
rd = Ridge(alpha=1)
rd.fit(X_train_trans, y_train)
```

```python
print("training ", rd.score(X_train_trans, y_train))
print("testing ", rd.score(X_test_trans, y_test))
```

```python
from sklearn.linear_model import Lasso
laso = Lasso(alpha=0.1)
laso.fit(X_train_trans, y_train)
```

```python
print("Training: ", laso.score(X_train_trans, y_train))
print("Testing: ", laso.score(X_test_trans, y_test))
```
